##Projeto de Notas de Alunos
Instruções do projeto
Desenvolva um programa que deve ler um arquivo csv intitulado “notas_alunos.csv”. O arquivo deve ter a seguinte estrutura:

aluno: Nome do aluno;
nota_1: Primeira nota;
nota_2: Segunda nota;
faltas: Número de faltas;

O programa lerá esse arquivo e criará duas colunas. A primeira coluna será “media”, que terá a média das duas notas do aluno. A segunda será “situacao”, com os valores: APROVADO ou REPROVADO.

O aluno que tiver mais de 5 faltas ou possuir média menor que sete, será reprovado. O programa deverá salvar esse novo dataframe com o nome “alunos_situacao.csv”.

Por fim, o programa deverá mostrar na tela:
- o maior número de faltas;
- a média geral das notas dos alunos;
- e a maior média.

#### Importando biblioteca

```python
import pandas as pd
import os  
```

#### Importando o arquivo notas_alunos.csv

```python
df = pd.read_csv("notas_alunos.csv")
```

```python
#As primeiras linhas da tabela (dataframe)
df.head()
```

#### Criando a primeira coluna Média

```python
media = (df["nota_1"] + df["nota_2"]) / 2
```

```python
df["Média"]=media
```

```python
df.head()
```

#### Criando a coluna Situação

```python
df.loc[df["Média"]>=7,"Situação"] = "APROVADO"
df.loc[df["Média"]<7,"Situação"] = "REPROVADO"
```

```python
df.head()
```

```python
df.loc[df["faltas"]<=5,"Situação"] = "APROVADO"
df.loc[df["faltas"]>5,"Situação"] = "REPROVADO"
```

```python
df.head()
```

#### Salvando no arquivo alunos_situacao.csv

```python
os.makedirs('folder/subfolder', exist_ok=True)  
df.to_csv('folder/subfolder/alunos_situacao.csv') 
```

#### Gerando o relatório 
o maior número de faltas;
a média geral das notas dos alunos;
e a maior média.

```python
maiorFaltas = df["faltas"].max()
mediaGeral = df["Média"].median()
maiorMedia = df["Média"].max()
print("O maior número de faltas foi: ", maiorFaltas)
print("A média Geral foi de: ", mediaGeral)
print("A maior média foi: ", maiorMedia)
```
